```python
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
from sklearn.pipeline import Pipeline

from utils.processing import *
from recommender import * 
```

```python
def drop_cols_for_q2(df):
    drop_columns = [
        'address',
        'title',
        'listing_id',
        'property_name',
        'total_num_units',
        'available_unit_types',
        'property_details_url',
        'elevation',
        'tenure',
        'property_type',
        'floor_level',
        'furnishing',
        'built_year',
        ### 
        'subzone', 
        'planning_area', 
        'region'
    ]
    return df.drop(columns=drop_columns)
    
df = pd.read_csv('../data/train.csv')
df = preprocess(df)

adfs = read_aux_csv('../data')
df = join_aux(df, adfs)

df_with_listing_info = df.copy()
df = drop_cols_for_q2(df)
df.head(1)
```

```python
pipe = Pipeline([('scaler', StandardScaler())])
df = pipe.fit_transform(df)
```

```python
row_idx = 0
num_recommendations = 5 
```

```python
knn_model = knn()
knn_model.fit(df, n_neighbors=num_recommendations, algorithm='auto')
knn_model.get_top_recommendations(df_with_listing_info, row_idx)[['listing_id', 'title', 'property_type', 'price', 'size_sqft']]
```

```python
knn_model = knn()
knn_model.fit(df, n_neighbors=num_recommendations, algorithm='ball_tree')
knn_model.get_top_recommendations(df_with_listing_info, row_idx)[['listing_id', 'title', 'property_type', 'price', 'size_sqft']]
```
